```r
#Create a function so that we can clean this dataset in one go!
```

```r
readStates <- function(){
  x <- read.csv("https://www2.census.gov/programs-surveys/popest/tables/2010-2011/state/totals/nst-est2011-01.csv", stringsAsFactors = FALSE)
  x <- x[-c(1:8,60:66), -(6:10)]
  colnames(x) <- c("stateName", "Census",  "Estimated", "Pop2010", "Pop2011")
  x[,1] <- sub(".", "", x[,1])
  for (i in 2:5) {
    x[,i] <- as.numeric(gsub(",", "", x[,i]))
  }
  return(x)
}
```

```r
#Execute the function and read the dataset in 'states' dataframe
```

```r
states <- readStates() 
states
```

WHICH VISUALISATION IS BETTER?

BOXPLOT - BOXPLOT IS BEST AT DETECTING THE OUTLIERS. IT IS ALSO GOOD FOR DIRECTLY GIVING YOU THE APPROAXIMATE VALUES OF QUARTILES, MEDIAN AND RANGE OF THE DATA.

HISTOGRAM - HISTOGRAM IS USEFUL WHEN WE WANT TO SEE THE DISTRIBUTION OF THE DATA. WE CAN ANALYSE THE PATTERN IN DATA, CHECK IF THE DISTRIBUTION IS LEFT SKEWED, RIGHT SKEWED OR NORMAL. 

SO, BOTH HAVE THEIR OWN ADVANTAGES BUT I FEEL HISTOGRAM IS BETTER VISUALISATION THAN BOXPLOT AS IT EXPLAINS THE DISTRIBUTIONS OF THE DATA.
